```python
ratings = pd.read_csv("data/UserRatings_ByGenre_all.csv")
```

```python
clustering_tb = ratings.set_index(ratings.columns[0])
```

```python
clustering_tb[:5]
```

```python
for i in clustering_tb.columns:
    clustering_tb[i] = clustering_tb[i].map(lambda x: 10 if x > 10 else x)
```

```python
clustering_tb.describe()
```

```python
from sklearn.cluster import KMeans
```

```python
km_5 =  KMeans(n_clusters=5, 
                init='k-means++',
                random_state=0)
km_5.fit(clustering_tb)
```

```python
cluster_df = pd.DataFrame(km_5.cluster_centers_, columns=clustering_tb.columns)
```

```python
for i in range(len(cluster_df)):
    ss = cluster_df.iloc[i].sort_values()
    print("group",i, "top3:")
    print(np.round(ss[-20:],2),"\n")
    print("worst3",)
    print(np.round(ss[:3],2),"\n")
```

```python
Group_1 = [ "Military", "Sci-Fi"]
Group_2 = [' Comedy', ' Shounen', 'Action']
Group_3 = [' Thriller', 'Thriller', ' Adventure',' Super Power']
```

```python
anime_info = pd.read_csv("data/anime_cleaned.csv")
```

```python
anime_info.columns
```

```python
anime_info.sort_values("score", ascending = False)[['anime_id', 'title', 'rating', 'score']]
```

```python
def intersection(lst1, lst2): 
    return list(set(lst1) & set(lst2)) 
```

```python
Group_1_anime = []
Group_2_anime = []
Group_3_anime = []

for i in range(anime_info.shape[0]):
    try:
        this_genre = anime_info["genre"].iloc[i].split(",")
    except:
        print(anime_info["genre"].iloc[i])
    if len(intersection(this_genre, Group_1)) >=1 :
        Group_1_anime.append((anime_info["anime_id"].iloc[i], anime_info["title"].iloc[i], anime_info["score"].iloc[i]))
    if len(intersection(this_genre, Group_2)) >=2 :
        Group_2_anime.append((anime_info["anime_id"].iloc[i], anime_info["title"].iloc[i],  anime_info["score"].iloc[i]))
    if len(intersection(this_genre, Group_3)) >=2 :
        Group_3_anime.append((anime_info["anime_id"].iloc[i],anime_info["title"].iloc[i],  anime_info["score"].iloc[i]))
```

```python
Group_1_anime.sort(key=lambda x: x[2], reverse=True)
Group_2_anime.sort(key=lambda x: x[2], reverse=True)
Group_3_anime.sort(key=lambda x: x[2], reverse=True)
```

## Mapping to users

```python
user_group = pd.DataFrame(km_5.labels_, index = clustering_tb.index, columns=["group"])
```

```python
user_group[:5]
```

```python
user_info = pd.read_csv("data/users_cleaned.csv")
```

```python
user_info.columns
```

```python
user_InfoGroup = pd.merge(user_group, user_info[['username', 'user_id','gender', 'location', 'birth_date' ]],
                         left_index= True, right_on = 'username')
```

```python
user_InfoGroup["birth_date"] = user_InfoGroup["birth_date"].map(lambda x: pd.to_datetime(x.split(" ")[0]))
```

```python
user_InfoGroup["age"] = user_InfoGroup["birth_date"].map(lambda x: pd.to_datetime("2018-12-03").year - x.year)
```

```python
user_InfoGroup[:5]
```

```python
def q_1(array):
    return np.percentile(array, 25)
def q_3(array):
    return np.percentile(array, 75)
```

```python
user_InfoGroup.groupby("group")["age"].agg({"mean": np.mean, 
                                            "3rd_quantile": q_3, 
                                            "1st_quantile": q_1})
```

```python
user_InfoGroup.groupby(["group", "gender"])["gender"].count()
```
